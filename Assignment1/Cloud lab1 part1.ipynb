```python
# The code is for a 3 year degree registration process
print('\n****************Registration Start****************\n')


# Entering details
Name = input('Enter Your Name: ')
Semester = int(input("Enter Semester: "))
Roll_num = input("Enter Your Roll number: ")

print('-------------------------------')
min_course = 4

if Semester%2 == 0:
    year = int(Semester/2)
else:
    year = int(Semester/2) + 1


# open courses file
file = open('C:/Users/HP/Desktop/sem 5/cloud lab/Assignment1/courses.txt','r')
courses = [line.strip() for line in file]


# According to your semester gathers courses 
cour_aval = []
for x in courses:
    if x[2] == str(year) and x[0] in {Roll_num[0],'H','S','M'}:
        cour_aval.append(x)



# Ask user to enter minimum number of courses
cour_num_taken = int(input('Enter the number of courses you want to take (minimum 4): '))
print('------------------------------------------------------------')
while cour_num_taken < min_course:
    print('*******You have to take minimum 4 courses*******')
    cour_num_taken = int(input('Enter the number of courses you want to take (minimum 4): '))


    
# Displays the courses available for user
print('Select the courses from the list below and enter the index: ')
print('-----------------------------------------------------------')
for i in range(len(cour_aval)):
    print(str(i+1) + ": " + cour_aval[i])
print('-----------------------------------------------------------')


# user enter the courses they choose
cour_taken = []
index = []
for i in range(cour_num_taken):
    x = int(input('Enter the course: '))
    if x not in index:
        index.append(x)
        cour_taken.append(cour_aval[x-1])
    else:
        while x in index:
            print('Course already Taken, Enter another Course')
            x = int(input('Enter the course: '))
        cour_taken.append(cour_aval[x-1])

# Displays the courses user selected
print('--------------------')
print("You have selected : ")
print('--------------------')
print(*cour_taken, sep='\n')


# Save the user information and courses they have taken in a file
file = open('C:/Users/HP/Desktop/sem 5/cloud lab/Assignment1/student_info.txt','a')
file.write(Name + ", " + Roll_num + ", " + str(Semester) + ", ")
for i in range(len(cour_taken)):
    file.write(cour_taken[i] + ", ")
file.write('\n')
# print(Name + " " + Roll_num + " " + str(Semester) + " " + " ".join(cour_taken))
file.close()
print('\n****************Registration Completed****************\n')
```
